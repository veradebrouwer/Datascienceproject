```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from collections import Counter
dutch_stopwords = stopwords.words('dutch')
```

```python
#Build request to rechtspraak.nl

zoekterm = ''
datum_begin = ''
datum_eind = ''
maximaal_aantal = ''

url ="https://data.rechtspraak.nl/uitspraken/zoeken?type=uitspraak&date=2011-05-%2001&date=2011-05-30"
```

```python
#Create dataframe of the data
url = 'https://data.rechtspraak.nl/uitspraken/zoeken?type=uitspraak&date=2011-05-%2001&date=2011-05-30'

document = requests.get(url)

soup= BeautifulSoup(document.content,"lxml-xml")
entries = soup.find_all("entry")

list_of_dicts = []

for entry in entries:
    if len(entry.find("summary").text) != 0:
        dict_entry = {'ID': entry.find("id").text, 'Title': entry.find("title").text, 'Summary':entry.find("summary").text}
        list_of_dicts.append(dict_entry)
    
df = pd.DataFrame(list_of_dicts)
```

```python
#Extract date from Title and create datetime column
def title_date(title):
    x=re.search(r'\d{2}-\d{2}-\d{4}', title)
    return x.group()

df['Date'] = df['Title'].apply(title_date)
df['Date'] = pd.to_datetime(df['Date'])
```

```python
#remove stopwords from summary
df['Summary without stopwords'] = df['Summary'].apply(lambda x: ' '.join([word for word in x.lower().split() if word not in (dutch_stopwords)]))

#generate wordcloud for dataset
wordcloud = WordCloud().generate(str(df['Summary without stopwords'].values))

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
#Most occuring words
counters = Counter(str(df['Summary without stopwords'].values).strip().split())
most_occur = counters.most_common(50)
print(most_occur)
```
